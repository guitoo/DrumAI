```python
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
```

```python
def Ra(f):
    num = np.square(12194)*f**4
    den = (f**2+20.6**2)
    den *= np.sqrt((f**2+107.7**2)*(f**2+737.9**2))
    den *= (f**2+12194**2)
    return num/den
```

```python
fft_size=1023
FS=44100

f = np.arange(0,FS//2,FS/fft_size)
```

```python
gain = Ra(f)
```

```python
plt.plot(f, gain)
```

```python
gain = np.concatenate([gain, gain[-2:0:-1]])
if fft_size % 2 == 0:
    gain = gain * np.exp(1.j * np.pi*1/fft_size)
    imp = np.real(fft.ifft(gain))
    imp = np.roll(imp,fft_size//2-1)
else:
    imp = np.real(fft.ifft(gain))
    imp = np.roll(imp,fft_size//2)
```

```python
plt.plot(imp)
```

```python
from scipy import signal
```

```python
def loudness(f):
    square = signal.square(2*np.pi*f/FS*np.arange(10000))
    return (np.convolve(imp,square)**2).sum()
```

```python
freq = np.arange(1500,2500,10)
loudness_freq = np.copy(freq)
```

```python
for i in range(len(freq)):
    loudness_freq[i] = loudness(freq[i])
```

```python
plt.plot(freq, loudness_freq)
```
