```python
def beta(m, at):
    return np.cosh(1/(m-1) * np.arccosh(10**(at/20)))

def cheb (n, x):
    return np.where(np.less_equal(np.abs(x),1),
                    np.cos(n * np.arccos(x)),
                    np.where(np.less(x,0),
                             (1 - 2 * (n % 2)) * np.cosh(n * np.arccosh(-x)),
                             np.cosh(n * np.arccosh(x))))

def cheb(n, x):
    return np.where(np.less_equal(np.abs(x),1),
                    np.cos(n * np.arccos(x)),
                    ((x-np.sqrt(x**2-1))**n+(x+np.sqrt(x**2-1))**n)/2)


def chebwin(m, at):
    k = np.arange(0,m)
    b = beta(m,at)
    
    x = b*np.cos(np.pi*k/m)
        
    W = cheb(m-1, x)
    if m % 2:
        w = np.real(fft(W))
        w = fftshift(w)
    else:
        W = W * np.exp(1.j * np.pi*k/m)
        w = np.real(fft(W))
        w = np.roll(w,m//2-1)
    return w/max(w)
```

```python
from ddsp.core import linear_lookup, tf_float32
from tensorflow.keras.layers import Layer, RNN
from ddsp import processors
from ddsp.processors import Processor, TensorDict
```

```python
def delay_line(delay, audio):
    return tf.linalg.diag_part(linear_lookup(delay,audio))
```

```python
class ProcessorCell(processors.Processor):
    
    def call(self, inputs: TensorDict, states) -> tf.Tensor:

        controls = self.get_controls(**inputs)
        signal = self.get_signal(**controls, states=states)
        return signal
```

```python
#from tensorflow.keras.layers import Wrapper

class RNNProcessor(processors.Processor):
    
    def __init__(self, cell, name, trainable=False):
        self._layer = RNN(cell, return_sequences=True)
        super().__init__(name=name, trainable=trainable)
    
    def get_signal(self, **control):
        return self._layer(control)
```

```python
#@gin.register
class FeedbackDelayNetworkCell(ProcessorCell):
    
    """feedback delay network
    
    audio shape [n_channels]
    delay shape [n_channels]
    matrix shape [n_channels, n_channels]
    states shape [n_channels, max_delay]
    """
    
    def __init__(self,
               trainable=False,
               n_channels=3,
               max_delay=50,
               name='FDN_cell'):
        super().__init__(name=name, trainable=trainable)
        self._n_channels = n_channels
        self._max_delay = max_delay
        self.state_size = n_channels * max_delay
    
    def build(self, unused_input_shape):
   
        if self.trainable:
            initializer = tf.random_normal_initializer(mean=0, stddev=1e-6)
            self._matrix = self.add_weight(
                name='mixing_matrix',
                shape=[self._n_channels, self._n_channels],
                dtype=tf.float32,
                initializer=initializer)
            self._delay = self.add_weight(
                name='delays',
                shape=[1,self._n_channels],
                dtype=tf.float32,
                initializer=initializer)
        self.built = True
        
    def get_controls(self, audio, delay=None, matrix=None):
        if self.trainable:
            delay = self._delay
            matrix = self._matrix
        else:
            if delay is None:
                raise ValueError('Must provide "delay" tensor if FDN trainable=False.')
            if matrix is None:
                raise ValueError('Must provide "matrix" tensor if FDN trainable=False.')

        matrix = tf.reshape(matrix, [n_channels, n_channels])
        return {'audio': audio, 'delay': delay, 'matrix': matrix}
    
    def get_signal(self, audio, delay, matrix, states):
        
        audio, delay, matrix = tf_float32(audio),tf_float32(delay),tf_float32(matrix)
        states = tf.reshape(states, [n_channels,max_delay])
        delayed = delay_line(delay,states)
        audio = delayed + audio
        mixed = tf.keras.backend.dot(audio, matrix)
        
        states = tf.slice(states, [0,1], [n_channels,max_delay-1])
        states = tf.concat([states, tf.transpose(mixed)], axis=1)
        audio = tf.math.reduce_sum(mixed, axis=1)
        
        return audio, [tf.reshape(states,[1,n_channels*max_delay])]

# def FeedbackDelayNetwork(n_channels=3, max_delay=50, name="FDN", trainable=False):
#     layer = RNN(FeedbackDelayNetworkCell(n_channels=n_channels, max_delay=max_delay, trainable=trainable), return_sequences=True)
#     params = ['audio', 'delay', 'matrix']
#     return RNNProcessor(layer, params, name)
```

```python
from tensorflow.keras import Input, Model

n_channels = 16
max_delay = 10

sound, sr = librosa.load(soundfiles[1])
sound = sound[:64]

signal_length = len(sound)
impulse_length = 3

imp_array = [1/2+np.cos(2*np.pi*x/impulse_length)/2 for x in range(-impulse_length//2, impulse_length//2)]

imp0 = tf.convert_to_tensor(imp_array, dtype='float32')
imp0 = tf.reshape(imp0, shape=[1, impulse_length, 1])
imp1 = tf.constant(0, shape=[1, signal_length-impulse_length, 1], dtype='float32')
impulse = tf.concat([imp0, imp1], axis=1)

fdn = FeedbackDelayNetwork(n_channels=n_channels, max_delay=max_delay, trainable=True)
spectral_loss = ddsp.losses.SpectralLoss(fft_sizes=(512, 256, 128, 64))
spectral_loss.__name__ = 'spectral_loss'

rms_loss = tf.keras.losses.MeanSquaredError()

impulse_input = Input(shape=[signal_length,1])
output_sound = fdn(impulse_input)

#spectral_loss = ddsp.losses.SpectralLoss(input_sound, output_sound)
model = Model(inputs=impulse_input, outputs=output_sound)

model.compile(optimizer= 'adamax', loss=spectral_loss)
```

```python
sound = sound.reshape(1,-1,1)
model.fit(impulse, sound, batch_size=1, epochs=100)
```
